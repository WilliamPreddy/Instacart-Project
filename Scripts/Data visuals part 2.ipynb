# Import Libraries and Pickle

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Import pickle
df= pd.read_pickle(r'C:\Users\william\5-2023 Instacart Basket Analysis\Data\Prepared Data\ords_prods_customers_merged.pkl')
```

# Creating Subset

```python
# Subset creation
df_subset = df[:1000000]
```

# Bar Chart

```python
# Bar chart for what days have the most orders
bar=df_subset['orders_days_of_the_week'].value_counts().sort_index().plot.bar()
```

# Histogram

```python
# Histogram of range of prices
df_subset['prices'].plot.hist(bins = 100)
```

# Scatter Plot

```python
# Scatter plot of prices
sns.scatterplot(x = 'prices', y = 'prices',data = df_subset)
```

# Line Chart

```python
# Line chart correlation between when products are ordered and their prices
orders_and_prices_line_chart=sns.lineplot(data = df_subset, x = 'orders_days_of_the_week',y = 'prices')
```

# Histogram

```python
# Histogram of when things are ordered
df_subset['time_ordered'].plot.hist(bins = 100)
```

```python
histogram_hour_ordered=df_subset['time_ordered'].plot.hist(bins = 100)
```
